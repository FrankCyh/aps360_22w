### Part (a) Convolutional Network - 5 pt

Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture
letter. Your model should be a subclass of nn.Module. Explain your choice of neural network architecture: how
many layers did you choose? What types of layers did you use? Were they fully-connected or convolutional?
What about other decisions like pooling layers, activation functions, number of channels / hidden units?

```python
class CNN_classifier(nn.Module):
    def __init__(self, kernel_sizes=[10, 5, 3], name="CNN_classifier"):
        super(CNN_classifier, self).__init__()
        # convolutional layers
        self.conv1 = nn.Conv2d(3, 5, kernel_sizes[0])
        self.conv2 = nn.Conv2d(5, 10, kernel_sizes[1])
        self.conv3 = nn.Conv2d(10, 25, kernel_sizes[2])
        # pooling layer
        self.pool = nn.MaxPool2d(2, 2)
        # Compute correct input size of the Fully Connected Layer
        self.name = name
        self.x = floor(0.5 * (224 + 1 - kernel_sizes[0]))
        self.z = floor(0.5 * (self.y + 1 - kernel_sizes[2]))
        self.y = floor(0.5 * (self.x + 1 - kernel_sizes[1]))
        self.FC_input = 25 * self.z * self.z
        self.fc1 = nn.Linear(self.FC_input, 32)
        self.fc2 = nn.Linear(32, 9)

    def forward(self, img):
        x = self.pool(F.relu(self.conv1(img)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = x.view(-1, self.FC_input)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
def train(model, train_dataset, val_dataset, batch_size=128, num_epochs=20, learn_rate=0.001):
    torch.manual_seed(21)
    train_acc = np.zeros(num_epochs)
    val_acc = np.zeros(num_epochs)

    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size)
    val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=batch_size)

    criterion = nn.CrossEntropyLoss() # loss choosed to be cross entropy loss
    # Reason: entropy loss is good for multi-class classification problem
    optimizer = optim.Adam(model.parameters(), lr=learn_rate) #optimizer is Adam with momentum
    # Reason: adam is training with momentum, thus better than SGD

    ## Train
    n = 0
    for epoch in range(num_epochs):
        for imgs, labels in iter(train_loader):

            if use_cuda and torch.cuda.is_available():
                imgs = imgs.cuda()
                labels = labels.cuda()

            out = model(imgs)
            loss = criterion(out, labels)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            n += 1

        # record accuracy
        train_acc[epoch] = get_accuracy(model, train_loader)
        val_acc[epoch] = get_accuracy(model, val_loader)
        print(f"Epoch {epoch}: Train accuracy: {train_acc[epoch]} |" + "Validation accuracy: {val_acc[epoch]}")

        # save model
        model_path = get_model_name(model.name, batch_size, learn_rate, epoch)
        torch.save(model.state_dict(), model_path)

    epochs = np.arange(1, num_epochs + 1)

    return train_acc, val_acc, epochs
```

```python
def get_model_name(name, batch_size, learning_rate, epoch):
    path = f"model_{name}_bs{batch_size}_lr{learning_rate}_epoch{epoch}"
    return path
```

```python
def get_accuracy(model, data_loader):
    tot_sample = 0
    correct = 0
    for imgs, labels in data_loader:
        tot_sample += imgs.shape[0]
        output = model(imgs)
        pred = output.max(1)[1]
        correct += pred.eq(labels.view_as(pred)).sum().item()
    return correct / tot_sample
```

The alexnet model is split up into two components: *alexnet.features* and *alexnet.classifier*. The

The neural network alexnet.features expects an image tensor of shape Nx3x224x224 as input and it will
